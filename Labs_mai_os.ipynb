Условие:
> Написать скрипт на языке Python, имитирующий работу записной книжки, обеспечить следующий функционал:
* При запуске программы необходимо вывести содержимое записной книжки.
* Потом пользователь вводит записи по-строчно.
* После ввода команды "Stop" - ввод завершается, введенные строки сортируются в алфавитном порядке и записываются в файл.

При решении этой задачи можно обойтись без использования библиотек, пускай пользователь вводит имя файла - блокнота самостоятельно.

Напишем функцию ***notebook(filename)***, которая принимает в качестве аргумента имя файла - записной книжки.

````python
def notebook(filename):
  with open(filename.strip(), 'a+', encoding='Utf-8') as file:
    current = file.read()
    strings = []
    if current:
      print(f'File \'{filename}\' has such content:\n{current}\n')
    else:
      print(f'File {filename} is empty!\n')
    s = ''
    print('Please, enter your messages, type \'Stop\' if you want to stop it\n')
    while s != 'Stop':
      s = input()
      strings.append(s)
    file.write('\n'.join(sorted(strings[:-1], key=lambda x: x[0])) + '\n')


if __name__ == '__main__':
  filename = input('Please, enter the name of a file: ')
  notebook(filename)
````

**Вывод:** В результате выполнения лабораторной работы был написан скрипт на языке Python, работающий в формате записной книжки.

### IPC

#### Реализация мульти-нитей(multithreading)

##### Использование os.fork(в Unix подобных системах) или *os.spawnve*(в Windows)

Функция **os.fork/os.spawnve** $-$ создаёт практически полную копию процесса-родителя, выполняещего вызов этой функции. 

После вызова этой функции алгоритм обычно разветвляется, а функция возвращает одно из 3-х значенией:
* (-1) - если вызов закончился неудачно
* (0) - если вызов удался, то значение идет процессу-родителю
* (PID) - если вызов удался, то значение идет процессу-потомку

Пример:

````python
import os
import sys
````

````python
pid = os.spawnv(mode=os.P_NOWAIT, file=sys.executable, args=[sys.executable])

if pid == -1:
  print('Could not start a new process')
elif pid > 0:
  try:
    os.execl(sys.executable, sys.executable, 'alpha')
    print("Process ID:", os.getpid())
    print("Child's process ID:", pid)
  except Exception as ex:
    print(ex)

else:
  os.exel(sys.executable, sys.executable, 'alpha')   
  print("Process ID:", os.getpid())
  print("Child's process ID:", pid)
````

##### Создание дочернего класса от класса - родителя Threading

````python
import threading
from datetime import datetime as dt
````

````python
class CountingThread(threading.Thread):
    def __init__(self, filename):
        super(CountingThread, self).__init__()
        self.file = filename
    def run(self):
        with open(self.file, 'r', encoding='utf-8') as file:
          sleep(3)
          print(f'Other thread: <{dt.now()}: {threading.get_ident()}>')
          [[print(j) for j in i.split()] for i in file.readlines()]
        print('\nThreading done!')

print(f'Main thread: <{dt.now()}: {threading.get_ident()}>\nWorking....')

t1 = CountingThread('/home/test.txt')
t1.start()
t1.join()
````

#### Реализация мультипроцессов


````python
import multiprocessing
from datetime import datetime as dt
from time import sleep
````

````python
def run(file, delay):
  with open(file, 'r', encoding='utf-8') as file:
    sleep(delay)
    print(f'Starting process <{dt.now()}: ' \
          f'{multiprocessing.current_process().name}>')
    [[print(j) for j in i.split()] for i in file.readlines()]
  print('\nProcess done!')


p1 = multiprocessing.Process(target=run, args=('/home/test.txt', 3))
p2 = multiprocessing.Process(target=run, args=('/home/test2.txt', 5))
print(f'Starting processes from <{multiprocessing.current_process().name}>' \
      f' at time <{dt.now()}>')
p1.start(); p2.start()
p1.join(); p2.join()
````

````python
import time

def run(string, delay, num):
  process_name = multiprocessing.current_process().name
  print(f'\nStarting process <{dt.now()}: ' \
        f'{multiprocessing.current_process().name}>\n')
  print(f'Current phraze {num + 1}:')
  for i in string.split():
    print(f'{i}')
  time.sleep(delay)
  print(f'\nProcess <{process_name}> done at <{dt.now()}>')

processes = []
phrazes = ['Ride till i am done', 'Take ny horse to the old town road', 
           'Welcome to the hotel California', 'if I could save time ina bottle']
for i in range(4): 
  processes.append(
    multiprocessing.Process(target=run, args=(phrazes[i], 5.0, i)))
    

for p in processes: 
  p.start()
    
for p in processes: 
  p.join()
````

#### Системные прерывания


##### От клавиатуры


````python
from time import sleep
from threading import Thread
from threading import get_ident
from datetime import datetime as dt
from _thread import interrupt_main
import sys
````

````python
 # task executed in a new thread
def task(string: str, delay: int) -> None:
    print(f'Other thread: <{dt.now()}: {get_ident()}>\nStarting...')
    for num, val in enumerate(string.split()):
      if num == 1:
        sleep(delay)
        interrupt_main()
 

print(f'Main thread: <{dt.now()}: {get_ident()}>\nWorking....')
thread = Thread(target=task('I will kill you', 3))
thread.start()
try:
    while True:
        print('Main thread waiting...')
        sleep(0.5)
except KeyboardInterrupt:
    print('Main interrupted! Exiting.')
    # exit(0)
````

Напишем скрипт для реализации 5 процессов, в каждом по 3 нити, а также процесс безопасный журнал(для обеспечения безопасности будем использовать очередь)

````python
import logging
import sys
from time import sleep
from random import random
from logging.handlers import QueueHandler
from datetime import datetime as dt
from threading import current_thread, Thread, get_ident
from multiprocessing import current_process, Process, Queue


def logger_process(queue):
    logger = logging.getLogger()
    if logger.hasHandlers():
        logger.handlers.clear()
    logger.addHandler(logging.StreamHandler(sys.stderr))
    logger.setLevel(logging.DEBUG)
    while True:
        message = queue.get()
        if message is None:
            break
        logger.handle(message)


def thread_task(string):
    string = string.split()
    logger = logging.getLogger()
    if logger.hasHandlers():
        logger.handlers.clear()
    handler = logging.StreamHandler(sys.stderr)
    handler.setFormatter(logging.Formatter(fmt='[%(asctime)s] [%(name)s] [%(levelname)s] > %(message)s',
                                           datefmt='%Y-%m-%d %H:%M:%S'))
    logger.addHandler(handler)
    logger.setLevel(logging.DEBUG)
    logger.info(f'Child Process: <{current_process().name}> thread <{get_ident()}> \
    started.')
    for i in string:
        logger.debug(f'Child {current_process().name} thread <{get_ident()}> word <{i}>.')
        sleep(random())
    logger.info(f'Child Process: <{current_process().name}> thread <{get_ident()}> \
    ended.')


def task(string):
    logger = logging.getLogger()
    if logger.hasHandlers():
        logger.handlers.clear()
    handler = logging.StreamHandler(sys.stdout)
    handler.setFormatter(logging.Formatter(fmt='[%(asctime)s] [%(name)s] [%(levelname)s] > %(message)s',
                                           datefmt='%Y-%m-%d %H:%M:%S'))
    logger.addHandler(handler)
    logger.setLevel(logging.DEBUG)
    process = current_process()
    logger.info(f'Child Process: <{process.name}> \
    started.')
    threads = [Thread(target=thread_task, args=(string,)) for _ in range(3)]
    for thread in threads:
        thread.start()
    for thread in threads:
        thread.join()
    logger.info(f'Child Process: <{process.name}> \
    ended.')


if __name__ == '__main__':
    queue = Queue()
    logger = logging.getLogger()
    if logger.hasHandlers():
        logger.handlers.clear()
    handler = logging.StreamHandler(sys.stdout)
    handler.setFormatter(logging.Formatter(fmt='[%(asctime)s] [%(name)s] [%(levelname)s] > %(message)s',
                                           datefmt='%Y-%m-%d %H:%M:%S'))
    logger.addHandler(handler)
    logger.setLevel(logging.DEBUG)
    logger_p = Process(target=logger_process, args=(queue,), daemon=True)
    logger_p.start()
    logger.debug(f'<{current_process().name}> started')
    processes = [Process(target=task, args=('Sweet home alabama',)) for i in range(5)]
    for process in processes:
        process.start()
    for process in processes:
        process.join()
    logger.debug(f'<{current_process().name}> ended.')
    queue.put(None)
````
